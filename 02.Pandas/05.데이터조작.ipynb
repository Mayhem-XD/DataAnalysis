- 정렬

```python
np.random.seed = 2023
df = pd.DataFrame(np.random.randint(40,101,size=(4,5)),index=list('abcd'),columns=list('VWXYZ'))
df
```

```python
df.sort_values(by='V',ascending=False)
```

```python
# V 열에 동일한 값이 있으면, 동일한 값은 W 열을 보고 정렬
df.sort_values(by=['V','W'],ascending=False)
```

```python
df2 = df.sort_values(by=['V','W'],ascending=False)
df2.sort_index()
```

- 행/열의 합계

```python
df
```

```python
# 열의 합계(행 방향?)
df.sum(axis=0)
```

```python
# 행의 합계(열 방향?)
df.sum(axis=1)
```

```python
df['총점'] = df.sum(axis=1)
df
```

```python
df['평균'] = df[['V','W','X','Y','Z']].mean(axis=1)         # 리스트로 하는게 정석 
df
```

```python
df['평균'] = df.loc[:,:'d'].mean(axis=1)
df
```

```python
# 행 추가 방법
df.loc['합계',:] = df.sum()
df
```

- apply 변환

```python
# lambda 함수를 타이타닉 각 열에 적용
titanic.apply(lambda x: x.count())
```

```python
df2.sort_index()
```

```python
# 각 열에서 최대값 - 최소값 한 결과
df2.apply(lambda x: x.max()-x.min())    # x 는 각 col을 의미
```

```python
# V열의 제곱수
df2.V.apply(lambda x:x*x)           # x는 한 컬럼의 원소를 의미
```

```python
# 타이타닉 승객중 20살 기준 adult child 구별하는 라벨 열 추가
titanic['adult/child'] = titanic.age.apply(lambda x: 'adult' if x >=20 else 'child')
titanic.tail()
```

```python
data = {
    'price':['1,200','2,500','3,300'],
    'text':['03.입출력.ipynb','04.고급인덱싱.ipynb','05.데이터조작.ipynb']
    }
df3 = pd.DataFrame(data)
df3
```

```python
int('1,200'.replace(',',''))
```

```python
df3['정수'] = df3.price.apply(lambda x: int(x.replace(',','')))
df3
```

```python
df3['구둣점제거'] = df3.text.apply(lambda x: x.replace('.',''))
df3
```

```python
df3['한글제거'] = df3.구둣점제거.apply(lambda x: x.replace('ㄱ-ㅎㅏ-ㅣ가-히',''))
df3
```
