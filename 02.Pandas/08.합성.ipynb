### 6. 합성

```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])
df1
```

```python
df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2
```

```python
# Inner join - 컬럼명이 같은 열(여기선 고객번호)을 기준으로, 공통된 데이터만 합성
pd.merge(df1, df2)
```

```python
# Full outer join
pd.merge(df1,df2,how='outer')
```

```python
# Left outer join
pd.merge(df1,df2,how='left')
```

```python
# Right outer join
pd.merge(df1,df2,how='right')
```

- join method

```python
# 인덱스로 조인
df1.set_index('고객번호',inplace=True)
df2.set_index('고객번호',inplace=True)
```

```python
df1.join(df2)       # Left outer join
```

```python
pd.merge(df1,df2,left_index=True,right_index=True)
```

- concat

```python
# 연습문제 2
df6 = pd.DataFrame(np.random.randint(500,1000,size=(6,2))*1000, columns=['매출','비용'],index=list('123456'))
df6['이익'] = df6.매출 - df6.비용
df6
```

```python
df7 = pd.DataFrame(np.random.randint(500,1000,size=(6,2))*1000, columns=['매출','비용'],index=['7','8','9','10','11','12'])
df7['이익'] = df7.매출 - df7.비용
df7
```

```python
df8 = pd.concat([df6,df7])
df8
```

```python
df8.sum()
df8.loc['총실적'] = df8.sum()
df8
```
