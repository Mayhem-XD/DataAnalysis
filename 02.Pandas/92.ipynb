```python
import numpy as np
import pandas as pd
```

```python
filename = 'data/주민등록인구집계현황.csv'
df = pd.read_csv(filename, encoding='euc-kr',thousands=',')
df.head()
```

```python
df.isna().sum().sum()
```

```python
df['인구소멸 위기지역'] = (df.iloc[:,31:33].sum(axis=1)/ df.iloc[:,11:16].sum(axis=1)).round(2)
df.head()
```

```python
df = df[['행정구역구분명','행정구역명','인구소멸 위기지역']]
df.head()
```

```python
# 수원시, 용인시등 구가 있는 시는 제거
gu_city = ['수원시', '고양시' ,'용인시', '성남시', '안산시', '안양시']
df.행정구역명 = df.행정구역명.str.strip()
drop_index = []
for index in df[df.행정구역구분명=='시군'].index:
    if df.행정구역명[index].split()[-1] in gu_city:
        drop_index.append(index)
drop_index
```

```python
df.drop(drop_index,inplace=True)
df.head()
```

```python
df = df[df.행정구역구분명.str.contains('시군|구')]
df.head()
```

```python
df.sort_values(by='인구소멸 위기지역').head(10)
```
