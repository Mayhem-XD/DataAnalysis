```python
import numpy as np
import pandas as pd
```

```python
filename = 'data/주민등록인구집계현황.csv'
df = pd.read_csv(filename, encoding='euc-kr',thousands=',')
df.head()
```

```python
df['인구소멸 위기지역'] = (df.iloc[:,31:33].sum(axis=1)/ df.iloc[:,11:16].sum(axis=1)).round(2)
```

```python
df.head()
```

```python
df1 = df[df.행정구역명.str.contains('수원시|고양시|용인시')]
df1.head()
```

```python
df1.pivot_table('인구소멸 위기지역','행정구역명',aggfunc='max').sort_values(by='인구소멸 위기지역',ascending=False).head(10)
```

```python
df2 = df[df.행정구역명.str.contains('성남시|안산시|안양시')]
df2.head()
```

```python
df2.pivot_table('인구소멸 위기지역','행정구역명',aggfunc='max').sort_values(by='인구소멸 위기지역',ascending=False).head(10)
```
