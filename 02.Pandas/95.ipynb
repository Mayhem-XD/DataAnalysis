```python
import numpy as np
import pandas as pd
```

```python
filename = 'data/주민등록인구집계현황.csv'
df = pd.read_csv(filename, encoding='euc-kr',thousands=',')
df.head()
```

```python
df.isna().sum().sum()
```

```python
df.tail()
```

```python
df['가임기 여성인구수'] = df.iloc[:,31:33].sum(axis=1)
df.head()
```

```python
df.iloc[:,11:16].head()
```

```python
df['60세 이상 노령인구'] = df.iloc[:,11:16].sum(axis=1)
df.head()
```

```python
df['인구소멸 위기지역'] =  (df['가임기 여성인구수']/df['60세 이상 노령인구'])
df.tail()
```

```python
df.dropna(subset=['인구소멸 위기지역'],inplace=True)
df.tail()
```

```python
# 20_39 여성수 / 60세 이상 노인 인구
cols = '행정구역구분명,행정구역명,총 인구수,인구소멸 위기지역'.split(',')
df = df[cols]
df
```
