```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/서울시인구.txt',sep='\t', thousands=',', skiprows=2)
df.head()
```

```python
df.isna().sum().sum()
```

```python
cols = '자치구 계 계.1 계.2  65세이상고령자'.split()
df = df[cols]
df.head()
```

```python
df = df.rename(columns={'자치구': '구별','계': '인구수','계.1': '내국인','계.2': '외국인','65세이상고령자': '고령자'})
df.head()
```

```python
df.drop([df.index[0]],inplace=True)
df.head()
```

```python
# df.columns = ['':'']
# 천단위 구분기호 없애고 정수로 변환
# for column in df.columns[1:]:
#     df[column] = df[column].str.replace(',','').astype(int)
```

```python
df['외국인비율'] = (df.외국인/df.인구수*100).round(2)
df['고령자비율'] = (df.고령자/df.인구수*100).round(2)
df.head()
```

```python
df.to_csv('data/pop.csv', index=False, encoding='euc-kr')
```
