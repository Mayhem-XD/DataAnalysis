```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/서울시인구.txt',sep='\t', thousands=',')
df.head()
```

```python
df.drop([df.index[0]],inplace=True)
df.head()
```

```python
df.drop([df.index[0]],inplace=True)
df.head()
```

```python
df.drop([df.index[0]],inplace=True)
df.head()
```

```python
temp_index = []
for i in range(len(df.index)):
    temp_index.append(i)
print(temp_index)
```

```python
df.index = temp_index
df.head()
```

```python
df = df.rename(columns={'자치구': '구별'})
df = df.rename(columns={'인구': '인구수'})
df = df.rename(columns={'인구.3': '내국인'})
df = df.rename(columns={'인구.6': '외국인'})
df.head()
```

```python
df.인구수 = df.인구수.str.replace(',','').astype(int)
df.외국인 = df.외국인.str.replace(',','').astype(int)
df['65세이상고령자'] = df['65세이상고령자'].str.replace(',','').astype(int)
df.head()
```

```python
df['외국인비율'] = (df.외국인/df.인구수*100).round(2) 
df['고령자비율'] = (df['65세이상고령자']/df.인구수*100).round(2)
df.head() 
```

```python
df.drop(columns=['기간','세대','인구.1','인구.2','인구.4','인구.5','인구.7','인구.8','세대당인구','65세이상고령자'],inplace=True)
df.head()
```

```python
df.to_csv('data/pop.csv', index=False, encoding='euc-kr')
```
