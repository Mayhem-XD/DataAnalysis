### 2. 여러가지 플롯

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

- Bar Chart

```python
y,x = [2,3,1],[0,1,2]
plt.title('Bar Chart')
plt.bar(x,y)
plt.xticks(x,['가','나','다'])
plt.yticks(sorted(y))
plt.xlabel('가나다'), plt.ylabel('빈도수')
plt.show()
```

```python
s = pd.Series(np.random.rand(10),index=list('abcdefghij'))
plt.bar(s.index,s.values)               # df index,df.columns_values
plt.title('Bar Chart')
plt.yticks([0,0.2,0.4,0.6,0.8]),plt.xticks([]), plt.grid()
plt.show()
```

```python
s.plot(kind='bar')
plt.show()
```

```python
df = pd.DataFrame(np.random.rand(20).reshape(10,2)
                ,index=list('abcdefghij'),
                columns=['M','N'])
plt.bar(df.index, df.M.values)               # df index,df.columns_values
plt.title('Bar Chart')
plt.show()
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic['class'].value_counts()
```

```python
s = titanic['class'].value_counts()
plt.figure(figsize=(4,2))
plt.barh(s.index,s.values)
plt.title('타이타닉호의 선실별 승객수')
plt.show()
```

-  Pie Chart

```python
ratio = [10,15,20,25]       # 합이 100 아니여도 ㄱㅊ
labers = ['감','귤','배','밤']
plt.pie(ratio, labels=labers,autopct='%.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.pie(ratio, labels=labers,autopct='%.1f%%',colors=colors,explode=explode,startangle=90,shadow=True)
plt.title('Pie Chart')
plt.show()
```

- 산점도

```python
np.random.seed(2023)
X, Y = np.random.normal(2,2,100),np.random.randn(100),           # 평균이 2 표준편차가 2
plt.figure(figsize=(5,3))
plt.scatter(X,Y)
plt.title('산점도')
plt.show()
```

```python
colors = np.random.rand(100)
X, Y = np.random.normal(2,2,100),np.random.randn(100),         
plt.figure(figsize=(5,3))
plt.scatter(X,Y,c=colors,alpha=0.5)
plt.title('색상을 준 산점도')
plt.show()
```

```python
size = 500 * np.random.rand(100)    
plt.figure(figsize=(5,3))
plt.scatter(X,Y,s=size,c=colors,alpha=0.5)
plt.title('Bubble Chart')
plt.show()
```
