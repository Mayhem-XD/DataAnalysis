- Anscombe's Quartet

```python
ans = sns.load_dataset('anscombe')
plt.figure(figsize=(10,8))
for index, roman in enumerate(ans.dataset.unique()):
    ds = ans[ans.dataset == roman]
    plt.subplot(2,2,index+1)
    plt.scatter(ds.x,ds.y)
    plt.plot([3,20],[4.5,13],'r')
    plt.title(f'Dataset {roman}'),plt.grid()
plt.show()
```

```python
ans = sns.load_dataset('anscombe')
plt.figure(figsize=(10,8))
for index, roman in enumerate(ans.dataset.unique()):
    ds = ans[ans.dataset == roman]
    plt.subplot(2,2,index+1)
    plt.ylim(2,14)
    plt.scatter(ds.x,ds.y)
    plt.plot([3,20],[4.5,13],'r')
    plt.title(f'Dataset {roman}'),plt.grid()
plt.show()
```

- subplots()

```python
ans = sns.load_dataset('anscombe')
fig , axes = plt.subplots(2,2,figsize=(10,8),sharex=True,sharey=True)
for index, roman in enumerate(ans.dataset.unique()):
    ds = ans[ans.dataset == roman]
    ax = axes[index//2,index%2]     # row index, columns index
    ax.scatter(ds.x,ds.y)
    ax.plot([3,20],[4.5,13],'r')
    ax.set_title(f'Dataset {roman}'),plt.grid()
fig.suptitle('Anscombe',fontsize=15)
plt.tight_layout()
plt.show()
```

- Axis 객체와 축

```python
fig, ax0 = plt.subplots()
ax1 = ax0.twinx()
ax0.set_title("2개의 y축 한 figure에서 사용하기")
ax0.plot([10, 5, 2, 9, 7], 'r-', label="y0")
ax0.set_ylabel("y0")
ax0.grid(False)
ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1")
ax1.set_ylabel("y1")
ax1.grid(False)
ax0.set_xlabel("공유되는 x축")
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
for color in ['red', 'green', 'blue']:
    n = 750
    x, y = np.random.rand(2, n)
    scale = 200.0 * np.random.rand(n)
    ax.scatter(x, y, c=color, s=scale, label=color,
            alpha=0.3, edgecolors='none')

ax.legend()
ax.grid(True)

plt.show()
```

- Iris 모든 품종의 특성에 대한 박스플롯을 하나의 그래프 안에 그리기

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.species.unique()
```

```python
fig , axes = plt.subplots(1,3,figsize=(9,9))
for index, spec in enumerate(iris.species.unique()):
    ax = axes[index]
    sl = iris[iris.species == spec].sepal_length
    sw = iris[iris.species == spec].sepal_width
    pl = iris[iris.species == spec].petal_length
    pw = iris[iris.species == spec].petal_width
    ax.boxplot((sl,sw,pl,pw), labels=['sepal length','sepal_width','petal_length','petal_width'])
    ax.set_title(f'Species {spec}')
fig.suptitle('Iris',fontsize=15)
plt.tight_layout()
plt.show()
```
