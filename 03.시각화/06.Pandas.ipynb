### 5. Pands 시각화

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

- Line graph

```python
np.random.seed(2023)
df =pd.DataFrame(np.random.randn(100, 3),columns=list('ABC'),index=pd.date_range('2023-04-01', periods=100)).cumsum()
df.head()
```

```python
df.plot(figsize=(5,3))
plt.title('plot method in Pandas')
plt.xlabel('date'),plt.ylabel('data');
```

- Bar Chart

```python
iris = sns.load_dataset('iris')
iris.sepal_length[:30].plot(kind='bar',rot=45,figsize=(6,3))
plt.title('Iris Setosa sepal_length');
```

```python
iris = sns.load_dataset('iris')
iris.sepal_length[:30].plot(kind='barh')
plt.title('Iris Setosa sepal_length');
```

```python
df2 = iris.groupby('species').mean()
df2
```

```python
df2.plot.bar(rot=0)
plt.title('avg');
```

```python
df2.columns.name = 'Feature'
df2.plot.bar(rot=0 ,figsize=(6,3))
plt.ylim(0,10)
plt.title('avg');
```

- Pie Chart

```python
titanic = sns.load_dataset('titanic')
s = titanic['class'].value_counts()
s
```

```python
s.plot.pie(autopct='%.1f%%')
plt.title('선실별 승객수');
```

- histogram, kernel density

```python
iris.plot.hist();
```

```python
iris.plot.kde();
```

- box plot

```python
iris[iris.species =='setosa'].plot.box();
```

```python
iris.boxplot(by='species')
plt.tight_layout()
plt.suptitle('각 피쳐의 품종별 박스 플롯');
```
