```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.embark_town.fillna('Southampton',inplace=True)
```

```python
sns.displot(df.survived,x=df.age)
plt.title('');
```

```python
df1 = df.pivot_table('survived','embark_town','sex').round(2)
df1
```

```python
sns.heatmap(data=df1,)
plt.title('embark_town,sex');
```

```python
sns.heatmap(df1, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('   ')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
filename = '../02.Pandas/data/   CCTV  _221231.csv'
colums = ',,2013   CCTV,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022'.split(',')
cctv = pd.read_csv(filename, encoding='euc-kr',thousands=',')
cctv=cctv[colums]
```

```python
cctv.drop([cctv.index[0]],inplace=True)
cols = ' '.split()
cctv = cctv[cols]
cctv.head()
```

```python
sns.barplot(x='',y='',data=cctv);
```

```python
cctv1 = cctv.sort_values(by='',ascending=True)
sns.barplot(x='',y='',data=cctv1);
```
