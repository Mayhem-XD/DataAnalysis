```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.embark_town.fillna('Southampton',inplace=True)
```

```python
sns.displot(df.survived,x=df.age,kde=True)
plt.title('');
```

```python
df1 = df.pivot_table('survived','embark_town','sex').round(2) * 100
df1
```

```python
sns.heatmap(data=df1,)
plt.title('embark_town,sex');
```

```python
sns.heatmap(df1, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('성별과 탑승지에 따른 생존률')
plt.xlabel('성별')
plt.ylabel('탑승지')
plt.show()
```

```python
filename = '../02.Pandas/data/서울시 자치구 년도별 CCTV 설치 현황_221231기준.csv'
colums = '구분,총계,2013년 이전 설치된 CCTV,2013년,2014년,2015년,2016년,2017년,2018년,2019년,2020년,2021년,2022년'.split(',')
cctv = pd.read_csv(filename, encoding='euc-kr',thousands=',')
cctv=cctv[colums]
```

```python
cctv.drop([cctv.index[0]],inplace=True)
cols = '구분 총계'.split()
cctv = cctv[cols]
cctv.head()
```

```python
sns.barplot(x='총계',y='구분',data=cctv);
```

```python
cctv1 = cctv.sort_values(by='구분',ascending=True)
sns.barplot(x='총계',y='구분',data=cctv1);
```

```python
cctv2 = cctv.sort_values(by='총계',ascending=True)
sns.barplot(x='총계',y='구분',data=cctv2);
```
