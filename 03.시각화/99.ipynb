```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

tips = sns.load_dataset('tips')
tips['tip_pct'] = (tips.tip / tips.total_bill * 100).round(2)
test = tips.pivot_table('tip_pct', 'day', aggfunc='mean').round(2)

plt.bar(test.index, test.values.flatten())

plt.title('요일별 팁 비율')
plt.xlabel('요일')
plt.ylabel('Tip percentage')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

filename = '../02.Pandas/data/서울시 자치구 년도별 CCTV 설치 현황_221231기준.csv'
colums = '구분,총계'.split(',')
cctv = pd.read_csv(filename, encoding='euc-kr',thousands=',')
cctv=cctv[colums]
cctv = cctv.rename(columns={'구분': '구별','총계': 'CCTV댓수'})
cctv.head()
```

```python
cctv.drop([cctv.index[0]],inplace=True)
cctv.head()
```

```python
plt.barh(cctv.구별,cctv.CCTV댓수)
plt.title('서울시 구별 cctv 댓수')
plt.show()
```

```python
df1 = cctv.sort_values(by='구별',ascending=True)
plt.barh(df1.구별,df1.CCTV댓수)
plt.title('서울시 구별 cctv 댓수')
plt.show()
```

```python
df2 = cctv.sort_values(by='CCTV댓수',ascending=False)
plt.barh(df2.구별,df2.CCTV댓수)
plt.title('서울시 구별 cctv 댓수')
plt.show()
```
