```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('data/20230425_YouTube랭킹.csv')
df.head(3)
```

```python
df_ = df.sort_values(by=['순위'])
df_.head()
```

```python
plt.barh(df_.채널명[:20],df_.구독자수[:20]);
```

```python
# 구독자 수 Top 20 - barh chart
plt.barh(df_.채널명[:20],df.구독자수[:20])
plt.show()
```

```python
df2 = df.sort_values(by=['조회수'],ascending=False)
df2.head()
```

```python
plt.barh(df2.채널명[:20],df2.조회수[:20]);
```

```python
df3 = df.sort_values(by=['비디오'],ascending=False)
df3.head()
```

```python
plt.barh(df3.채널명[:20],df3.구독자수[:20]);
```

```python
sns.barplot(x=df3.비디오[:20],y=df3.채널명[:20]);
```

```python
df4 = df.카테고리.value_counts().to_frame()
```

```python
df4
```

```python
df.pivot_table('채널명','카테고리',aggfunc='count').sort_values('채널명',ascending=False)
```

```python
df4.head()
```

```python
sns.barplot(y=df4.index[:10],x='count',data=df4.head(10))
plt.title('카테고리별 채널 수 Top 10');
```

```python
df5 = df.pivot_table('구독자수','카테고리',aggfunc=sum).sort_values('구독자수',ascending=False)
df5.head(10)
```

```python
plt.title('카테고리별 구독자 수 Top 10')
sns.barplot(y=df5.index[:10],x=df5.구독자수.head(10));
```
