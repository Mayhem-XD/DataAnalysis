```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sw_wc = STOPWORDS.copy()
print(sw_wc)
```

```python
len(sw_wc)
```

```python
with open('data/test_review.txt') as f:
    text = f.read()
```

```python
sw_wc.add('one')
```

```python
wc = WordCloud(
    background_color = 'white',
    max_words = 4000,
    stopwords = sw_wc,
).generate(text)
```

```python
plt.figure(figsize=(10,6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
count = 0
for key,val in wc.words_.items():
    print(f'{key}:\t{val}')
    count += 1
    if count >= 10:
        break
```
