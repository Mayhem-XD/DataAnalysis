```python
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sw_wc = STOPWORDS.copy()
print(sw_wc)
```

```python
len(sw_wc)
```

```python
with open('data/test_review.txt') as f:
    text = f.read()
```

```python
words = 'one movie film'.split()
for word in words:
    sw_wc.add(word)
```

```python
wc = WordCloud(
    background_color = 'white',
    max_words = 4000,
    stopwords = sw_wc,
).generate(text)
```

```python
plt.figure(figsize=(10,6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
count = 0
for key,val in wc.words_.items():
    print(f'{key}:\t{val}')
    count += 1
    if count >= 10:
        break
```

```python
mask = np.array(Image.open('data/sr_mask.png'))
plt.imshow(mask,cmap=plt.cm.gray,interpolation='bilinear')
plt.axis('off');
```

```python
wc = WordCloud(
    background_color = 'black',
    max_words = 2000,
    stopwords = sw_wc,
    mask=mask
).generate(text)
plt.figure(figsize=(16,9))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
