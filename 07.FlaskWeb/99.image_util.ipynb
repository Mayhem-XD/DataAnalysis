```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open('static/img/icu.jpg')
plt.imshow(img),plt.axis('off');
```

```python
np.array(img).shape
```

```python
h,w,_ = np.array(img).shape
```

```python
if h > w:
    width, height = w,w
else:
    width, height = h,h
```

```python
diff = abs(h - w) // 2
```

```python
# landscape
if w > h:
    final_img = np.array(img)[:,diff:diff+h,:]
# portrait
else:
    final_img = np.array(img)[diff:diff+w:]
```

```python
new_img = Image.fromarray(final_img)
plt.imshow(new_img),plt.axis('off');
```

```python
new_img.save('static/img/profile.png')
```
