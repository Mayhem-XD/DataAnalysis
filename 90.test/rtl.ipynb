```python
import requests,json,datetime
from urllib.parse import quote 
import pandas as pd
import numpy as np
```

```python
with open('data/sub_rt_lo_key.txt') as f:
    rt_lo_key = f.read()
```

```python
# http://swopenAPI.seoul.go.kr/api/subway/(인증키)/xml/realtimePosition/0/5/1호선
base_url = 'http://swopenAPI.seoul.go.kr/api/subway/'
params1 = f'{rt_lo_key}/json/realtimePosition/0/5/'
params2 = quote('1호선')
url = f'{base_url}{params1}{params2}'
```

```python
response = requests.get(url)
res = json.loads(response.text)
res
```
