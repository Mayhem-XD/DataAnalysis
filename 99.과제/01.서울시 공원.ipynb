```python
import json, folium, re
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('data/pop.csv')
df1.head()
```

```python
df2 = pd.read_csv('data/서울시 주요 공원현황.csv',encoding='euc-kr')
df2 = df2['공원명,면적,공원주소'.split(',')]
df2.columns = ['공원명','면적','주소']
df2.head(7)
```

```python
df2.면적[6]
df2.drop([6],inplace=True)
```

```python
df2.면적 = df2.면적.str.replace('총 ','')
df2.head(7)
```

```python
df2.dropna(subset=['면적'],inplace=True)
df2.isna().sum().sum()
```

```python
df2.면적 = df2.면적.apply(lambda x: re.split('[㎥㎡m]',x)[0]) 
```

```python
df2.면적 = df2.면적.astype(float).astype(int)
df2.drop([2],inplace=True)
df2.head()
```

```python
df2.to_csv('data/서울공원.csv',index=False)
```

```python
df2['구별'] = df2.주소.apply(lambda x:x.split()[1])
df2.tail()
```

```python
df2.구별[129] = '노원구'
df2.tail()
```

```python
df3 = df2.groupby('구별')['면적'].sum().reset_index()
df3
```

```python
df4 = pd.merge(df1, df3)
```

```python
df5 = pd.read_csv('data/서울공원.csv',index_col='구별')
df5.head()
```

```python
df5['인당면적'] = (df5['면적'] / df5['인구수']).round(2)
df5.head()
```

```python
file = 'data/seoul_geo_simple.json'
geo_str = json.load(open(file, encoding='utf-8'))
```

```python
def get_text_location(ger_str):
    gu_dict = {}
    for gu in geo_str['features']:
        for coord in gu['geometry']['coordinates']:
            geo = np.array(coord)
            gu_dict[gu['id']]= [np.mean(geo[:,1]),np.mean(geo[:,0])]
    return gu_dict
```

```python
gu_dict = get_text_location(geo_str)
gu_dict
```

```python
df_map = folium.Map(location=[37.55,126.98], zoom_start=11,tiles='Stamen Toner')
folium.Choropleth(
    geo_data=geo_str,
    data=df5.면적,           # 단계구분도로 보여줄 데이터
    columns=[df5.index,df5.면적],      # df에서 추출할 항목
    fill_color='PuRd',                  # Colormap
    key_on='feature.id'                # 지도에서 조인할 항목
).add_to(df_map)
for gu_name in df5.index:
    folium.map.Marker(
        location=gu_dict[gu_name],
        icon = folium.DivIcon(icon_size=(80,20), icon_anchor=(20,0),
            html=f'<div style="font-size: 10pt">{gu_name}</div>'
        )
    ).add_to(df_map)
title_html = '<h3 align="center" style="font-size:20px">자치구별 공원 면적</h3>'
df_map.get_root().html.add_child(folium.Element(title_html))
df_map
```

```python
df2_map = folium.Map(location=[37.55,126.98], zoom_start=11,tiles='Stamen Toner')
folium.Choropleth(
    geo_data=geo_str,
    data=df5.인당면적,           # 단계구분도로 보여줄 데이터
    columns=[df5.index,df5.인당면적],      # df에서 추출할 항목
    fill_color='PuRd',                  # Colormap
    key_on='feature.id'                # 지도에서 조인할 항목
).add_to(df2_map)
for gu_name in df5.index:
    folium.map.Marker(
        location=gu_dict[gu_name],
        icon = folium.DivIcon(icon_size=(80,20), icon_anchor=(20,0),
            html=f'<div style="font-size: 10pt">{gu_name}</div>'
        )
    ).add_to(df2_map)
title_html = '<h3 align="center" style="font-size:20px">인당 공원 면적</h3>'
df2_map.get_root().html.add_child(folium.Element(title_html))
df2_map
```
