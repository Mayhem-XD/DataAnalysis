```python
import json, folium, re
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('data/전국도시공원정보표준데이터.csv',encoding='euc-kr')
df1 = df1['공원명,소재지도로명주소,소재지지번주소,공원면적,제공기관명'.split(',')]
df1.columns = ['공원명','도로명주소','주소','면적','제공기관명']
df1.head(30).tail(10)
```

```python
df1.제공기관명.isna().sum()
```

```python
df1 =df1[df1.제공기관명.str.contains('경기도')]
df1.head(50).tail(10)
```

```python
df_test = df1[['공원명','면적','도로명주소','주소']]
df_test.head()
```

```python
df_test.주소 = df_test.주소.fillna(df_test.도로명주소)
df_test.head(50).tail(10)
```

```python
df_test.isna().sum()
```

```python
df_test['구별'] = df_test['주소'].str.split().str[1:3].str.join(' ')
df_test.head()
```

```python
df_test_1 = df_test.groupby('구별')['면적'].sum().reset_index()
df_test_1.head()
```

```python
df_test_2 = df_test_1[df_test_1.구별.str.contains('수원시|고양시|용인시|성남시|안산시|안양시')]
df_test_2 = df_test_2.groupby('구별')['면적'].sum().reset_index()
```

```python
df_test_2.head(30).tail(15)
```

```python
test_col = ~df_test_1.구별.str.contains('수원시|고양시|용인시|성남시|안산시|안양시')
df_test_3 = df_test_1[test_col]
df_test_3
```

```python
df_test_3.구별 = df_test_3.구별.apply(lambda x: x.split()[0])
df_test_3.head()
```

```python
df_test_3 = df_test_3.groupby('구별')['면적'].sum().reset_index()
df_test_3
```

```python
df_test_1.head(130).tail(10)
```

```python
df_result = pd.concat([df_test_3,df_test_2],ignore_index=True)
df_result.head(45).tail(15)
```

```python
df_result.to_csv('data/경기공원.csv',index=False)
pd.read_csv('data/경기공원.csv').tail()
```

```python
# 행정동명 인구수 파일에서 경기도 시군구 별 인구수 더해서 df 하나 만듬
# 공원csv 파일에서 제공기관명으로 쪼개서 경기도 공원 찾음
# 이렇게 저렇게 해서 합쳐서 끝
```

```python
df2 = pd.read_csv('data/주민등록인구집계현황.csv',encoding='euc-kr')
df2 = df2['행정구역구분명,행정구역명,총 인구수'.split(',')]
df2 = df2.rename(columns={'총 인구수':'인구수'})
df2.head()
```

```python
df2 =df2[df2.행정구역구분명.str.contains('시군|구')]
df2.head()
```

```python
gu_city = ['수원시', '고양시', '용인시', '성남시', '안산시', '안양시']
df2.행정구역명 = df2.행정구역명.str.strip()
drop_index = []
for index in df2[df2.행정구역구분명 == '시군'].index:
    if df2.행정구역명[index].split()[-1] in gu_city:
        drop_index.append(index)
df2.drop(drop_index, inplace=True)
df2.head()
```

```python
df2.drop([96],inplace=True)
```

```python
df2['구별'] = df2['행정구역명'].str.split().str[1:].str.join(' ')
df2.head()
```

```python
df2 = df2[['구별','인구수']]
df2.head()
```

```python
df2.to_csv('data/경기인구수.csv',index=False)
pd.read_csv('data/경기인구수.csv').head(10)
```

```python
df_new = pd.merge(df2,df_result)
df_new.to_csv('data/경기인구공원.csv',index=False)
pd.read_csv('data/경기인구공원.csv').head()
```

```python
df_new['인당면적'] = (df_new.면적 / df_new.인구수).round(2)
df_new.tail()
```

```python
df_new['시군구'] = df_new.구별.str.replace(' ','').str.strip()
df_new.set_index('시군구',inplace=True)
df_new.head()
```

```python
df_new.head(30).tail(15)
```

```python
file = 'data/경기도_geo_simple.json'
geo_str = json.load(open(file, encoding='utf-8'))
```

```python
geo_park_map = folium.Map(location=[37.55,126.98], zoom_start=10,tiles='Stamen Toner')
folium.Choropleth(
    geo_data=geo_str,
    data=df_new.면적,           
    columns=[df_new.index,df_new.면적],      
    fill_color='PuRd',                 
    key_on='properties.name'                
).add_to(geo_park_map)
title_html = '<h3 align="center" style="font-size:20px">자치구별 공원 면적</h3>'
geo_park_map.get_root().html.add_child(folium.Element(title_html))
geo_park_map
```

```python
ger_parkpp_map = folium.Map(location=[37.55,126.98], zoom_start=10,tiles='Stamen Toner')
folium.Choropleth(
    geo_data=geo_str,
    data=df_new.인당면적,           
    columns=[df_new.index,df_new.인당면적],      
    fill_color='PuBu',                 
    key_on='properties.name'                
).add_to(ger_parkpp_map)
title_html = '<h3 align="center" style="font-size:20px">자치구별 공원 인당면적</h3>'
ger_parkpp_map.get_root().html.add_child(folium.Element(title_html))
ger_parkpp_map
```
